```python
import matplotlib.pyplot as plt
import time
from random import randint

def insertionSort(A):
    for i in range(1,len(A)):
        key=A[i]
        j=i-1
        while j>=0 and A[j]>key:
            A[j+1]=A[j]
            j=j-1
        A[j+1]=key

def merge(A,p,q,r):
    nl=q-p+1
    nr=r-q
    left=[0]*nl
    right=[0]*nr
    for i in range(nl):
        left[i]=A[p+i]
    for j in range(nr):
        right[j]=A[q+1+j]
    a=0
    b=0
    c=p
    while a<nl and b<nr:
        if left[a]<=right[b]:
            A[c]=left[a]
            a+=1
        else:
            A[c]=right[b]
            b+=1
        c+=1
    while a<nl:
        A[c]=left[a]
        a+=1
        c+=1
    while b<nr:
        A[c]=right[b]
        b+=1
        c+=1

def mergeSort(A,p,r):
    if p>=r:
        return
    q=(p+r)//2
    mergeSort(A,p,q)
    mergeSort(A,q+1,r)
    merge(A,p,q,r)

len_of_array=[]
insertSort_tc=[]
mergeSort_tc=[]

for i in range(10):
  value=randint(100,1000)
  len_of_array.append(value)

len_of_array=sorted(len_of_array)

for value in len_of_array:
  arr1=[]
  for i in range(value):
    val=randint(1,value)
    arr1.append(val)

  start_is=time.time()
  insertionSort(arr1)
  end_is=time.time()
  time_is=round((end_is-start_is)*10**3,2)
  insertSort_tc.append(time_is)

  arr2=[]
  for i in range(value):
    val=randint(1,value)
    arr2.append(val)

  start_ms=time.time()
  mergeSort(arr1,0,len(arr1)-1)
  end_ms=time.time()
  time_ms=round((end_ms-start_ms)*10**3,2)
  mergeSort_tc.append(time_ms)

plt.plot(len_of_array,insertSort_tc,'r',label="Insertion Sort")
plt.plot(len_of_array,mergeSort_tc,'b',label="Merge Sort")

plt.xlabel("Length of array")
plt.ylabel("Time Complexity in milliseconds")
plt.legend()
plt.show()
```
